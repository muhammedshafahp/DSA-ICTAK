edible or poisonous (e or p) using the given dataset.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Data input

```python
data=pd.read_csv('/content/drive/MyDrive/DSA WORKS/week15/mushrooms.csv')
data
```

```python
data.isna().sum()
```

as though no null value we can see missind value with character '?' in the data we need to replace that

```python
data.describe()
```

```python
modevalue=data['stalk-root'].mode()
modevalue
```

Replacing the null data '?' in the data

```python
data1=data.replace(to_replace='?',value=modevalue)
```

```python
data1
```

Spliting data for modelling

```python
x=data1.drop('class',axis=1)
y=data1.iloc[:,1]
```

Encoding the modeleing data

```python
x=pd.get_dummies(x)
```

spliting data for train and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_train
```

Applying standardscaling to the data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
```

Applying Gaussian method

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
y_pred
```

Checking accuary and confusion matrixs

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred) 
```
