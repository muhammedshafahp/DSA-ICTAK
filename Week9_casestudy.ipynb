Check whether all the zones generate the same amount of sales

H0:Sales generated by all zones are equal.

H1:Sales generated by all zones are not equal.

```python
from scipy.stats import f_oneway
Fstatic,Pvalue=f_oneway(data['Zone - A'],data['Zone - B'],data['Zone - C'],data['Zone - D'])
print('F-static=',Fstatic,'P-value',Pvalue)
```

As the pvalue is less than 0.05 so that we can reject the Null Hypothesis, we can say that sales generated by each zone is different.
